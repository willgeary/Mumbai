```python
import geopandas as gpd
import matplotlib.pyplot as plt

%matplotlib inline
```

# Mumbai Wards

Source: https://github.com/mickeykedia/India-Maps

```python
mumbai_wards = gpd.read_file('../maps/mumbai/data/mumbai_wards/mumbai_wards.shp')
f, ax = plt.subplots(1, figsize=(8, 12))
ax = mumbai_wards.plot(ax=ax, color='lightgrey', linewidth=0.2, edgecolor='black')
ax.set_axis_off()
```

# Mumbai Outline

Dissolve the wards together into one vector.

```python
mumbai_outline = mumbai_wards[['name', 'shape_area', 'geometry']]
mumbai_outline = mumbai_outline.dissolve(by='name', aggfunc='sum')
f, ax = plt.subplots(1, figsize=(8, 12))
ax = mumbai_outline.plot(ax=ax, color='lightgrey', linewidth=0)
ax.set_axis_off()
mumbai_outline.to_file("../maps/mumbai/data/mumbai_outline/mumbai_outline.shp")
```

# Create Vector Grid

See https://gis.stackexchange.com/questions/54119/creating-square-grid-polygon-shapefile-with-python

```python
import shapefile as shp
import math
```

```python
def shapefile2grid(minx,maxx,miny,maxy, dx, dy, outputfile):
    nx = int(math.ceil(abs(maxx - minx)/dx))
    ny = int(math.ceil(abs(maxy - miny)/dy))

    w = shp.Writer(shp.POLYGON)
    w.autoBalance = 1
    w.field("ID")
    id=0

    for i in range(ny):
        for j in range(nx):
            id+=1
            vertices = []
            parts = []
            vertices.append([min(minx+dx*j,maxx),max(maxy-dy*i,miny)])
            vertices.append([min(minx+dx*(j+1),maxx),max(maxy-dy*i,miny)])
            vertices.append([min(minx+dx*(j+1),maxx),max(maxy-dy*(i+1),miny)])
            vertices.append([min(minx+dx*j,maxx),max(maxy-dy*(i+1),miny)])
            parts.append(vertices)
            w.poly(parts)
            w.record(id)

    w.save(outputfile)
```

```python
minx, maxx, miny, maxy = 8098833.37623,8124707.70179,2141821.35183,2187509.82486 # Mumbai extent from QGIS
dx, dy = 100, 100
shapefile2grid(minx, maxx, miny, maxy, dx, dy, "../maps/mumbai/data/mumbai_grid/mumbai_grid.shp")
```

# Vector Grid

```python
vector_grid = gpd.read_file('../maps/mumbai/data/mumbai_grid/vector_grid.shp')
f, ax = plt.subplots(1, figsize=(8, 12))
ax = vector_grid.plot(ax=ax, color='lightgrey', linewidth=0.2, edgecolor='black')
ax.set_axis_off()

print vector_grid.shape[0], "columns"
print vector_grid.shape[1], "columns"
```

# Gridded Mumbai

Here is the python code for an intersection set operation:

```python
mumbai_grid = gpd.overlay(mumbai_outline, vector_grid, how='intersection')
f, ax = plt.subplots(1, figsize=(8, 12))
ax = mumbai_grid.plot(ax=ax, color='lightgrey', linewidth=0.2, edgecolor='black')
ax.set_axis_off()
```
